```python
import random

days = list(range(1,32))
celsius = []
for _ in range(31):
    celsius.append(random.uniform(25,45))

#list comprehension
celsius1 = [random.uniform(25,45) for x in range(31) if x>25]
celsius1
```

```python
import random
import matplotlib.pyplot as plt

days = list(range(1,32))
#list comprehension 
celsius = [random.uniform(25,45) for _ in range(31)]

fig = plt.figure(figsize=(8,5))
ax = fig.add_subplot(111)
ax.plot(days,celsius,'c--*')
ax.grid(axis='y')
ax.set_xticks(days)
plt.show()
```
